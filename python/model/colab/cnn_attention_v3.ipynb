```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import random

from tensorflow.keras import layers, models, regularizers, optimizers, callbacks, mixed_precision
from tensorflow.keras.layers import (Dense, Conv2D, BatchNormalization, Activation, Input, Add, Flatten,
                                     GlobalAveragePooling2D, Concatenate, Reshape, SeparableConv2D,
                                     DepthwiseConv2D, AveragePooling2D, MaxPooling2D, Dropout)
from tensorflow.keras.regularizers import l2
from tensorflow.keras import backend as K

# Optional: Mixed precision for faster training (comment out if not needed)
mixed_precision.set_global_policy('mixed_float16')
tf.config.optimizer.set_jit(True)  # Enable XLA

###########################################################
# 1. Constants and Paths
###########################################################

TFRECORD_PATH = '/content/drive/MyDrive/ChessEvaluation/tfrecords'

# Hyperparameters
BATCH_SIZE = 1024
EPOCHS = 50
VALIDATION_SPLIT = 0.1
INITIAL_LEARNING_RATE = 1e-3  # Higher initial learning rate with decay
WEIGHT_DECAY = 1e-5           # L2 regularization

tfrecord_files = glob.glob(os.path.join(TFRECORD_PATH, '*.tfrecord'))
print(f"Found {len(tfrecord_files)} TFRecord files in {TFRECORD_PATH}")

# Shuffle TFRecord files before splitting
random.shuffle(tfrecord_files)

###########################################################
# 2. Data Loading and Parsing with Optimization
###########################################################

def parse_tfrecord_fn(example):
    """Parse a single TFRecord example into tensor features and labels."""
    feature_description = {
        'board_tensor': tf.io.FixedLenFeature([], tf.string),
        'extra_features': tf.io.FixedLenFeature([], tf.string),
        'evaluation': tf.io.FixedLenFeature([], tf.float32)
    }

    example = tf.io.parse_single_example(example, feature_description)

    # Parse the board representation (8x8x12)
    board = tf.io.parse_tensor(example['board_tensor'], out_type=tf.float32)
    board = tf.reshape(board, [8, 8, 12])

    # Parse extra features
    extra = tf.io.parse_tensor(example['extra_features'], out_type=tf.float32)
    extra = tf.ensure_shape(extra, [13])

    # Evaluation label
    evaluation = example['evaluation']

    # Data augmentation: flip the board horizontally with 50% probability during training
    # This creates a mirror position with the opposite evaluation
    if tf.random.uniform(()) > 0.5:
        board = tf.reverse(board, axis=[1])  # Flip horizontally
        evaluation = -evaluation  # Negate the evaluation

    return {'board': board, 'extra_features': extra}, evaluation


def create_dataset(file_pattern, batch_size=BATCH_SIZE, is_training=True):
    """Create a tf.data.Dataset from TFRecord files with optimized pipeline."""
    dataset = tf.data.TFRecordDataset(
        tf.io.gfile.glob(file_pattern),
        num_parallel_reads=tf.data.AUTOTUNE  # Parallelize file reading
    )

    # Parse records in parallel
    dataset = dataset.map(parse_tfrecord_fn, num_parallel_calls=tf.data.AUTOTUNE)

    if is_training:
        # Larger buffer for better shuffling
        dataset = dataset.shuffle(buffer_size=20000)

        # Repeat the dataset for multiple epochs
        dataset = dataset.repeat()

    # Batch and prefetch for performance
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)

    return dataset

###########################################################
# 3. Custom Loss Function with Temperature Scaling
###########################################################

def weighted_huber_loss(delta=0.1, focus_range=0.15, temp=10.0):
    """
    Huber loss with temperature scaling and focus weighting.

    Args:
        delta: Threshold for switching from quadratic to linear loss
        focus_range: Range around zero to emphasize
        temp: Temperature for scaling evaluations (higher = sharper training)
    """
    def loss(y_true, y_pred):
        # Optional: Apply temperature scaling to make model more confident
        # about clear positions while retaining precision in balanced positions
        scaled_true = tf.sign(y_true) * tf.pow(tf.abs(y_true), 1.0/temp)
        scaled_pred = tf.sign(y_pred) * tf.pow(tf.abs(y_pred), 1.0/temp)

        # Compute error on temperature-scaled values
        error = scaled_true - scaled_pred
        abs_error = tf.abs(error)

        # Huber loss: quadratic for small errors, linear for large ones
        quadratic = tf.minimum(abs_error, delta)
        linear = abs_error - quadratic
        huber_loss = 0.5 * tf.square(quadratic) + delta * linear

        # Apply focus weighting to emphasize balanced positions
        abs_values = tf.abs(y_true)
        weights = 1.0 + (5.0 * tf.maximum(0.0, (focus_range - abs_values)) / focus_range)

        return tf.reduce_mean(huber_loss * weights)

    return loss

###########################################################
# 4. Efficient MobileNetV2-inspired Chess Evaluation Model
###########################################################

def inverted_residual_block(x, filters, stride=1, expansion=6, dropout_rate=0):
    """
    Inverted residual block from MobileNetV2, adapted for chess evaluation.
    Uses depthwise separable convolutions for efficiency.
    """
    shortcut = x
    in_channels = K.int_shape(x)[-1]

    # Expansion phase
    if expansion > 1:
        x = Conv2D(
            expansion * in_channels,
            kernel_size=1,
            padding='same',
            use_bias=False,
            kernel_regularizer=l2(WEIGHT_DECAY)
        )(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

    # Depthwise convolution - note: kernel_regularizer is not supported in DepthwiseConv2D
    x = DepthwiseConv2D(
        kernel_size=3,
        strides=stride,
        padding='same',
        use_bias=False
        # Cannot use kernel_regularizer here - it's not supported
    )(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Projection phase
    x = Conv2D(
        filters,
        kernel_size=1,
        padding='same',
        use_bias=False,
        kernel_regularizer=l2(WEIGHT_DECAY)
    )(x)
    x = BatchNormalization()(x)

    # Optional dropout for regularization
    if dropout_rate > 0:
        x = Dropout(dropout_rate)(x)

    # Add shortcut if dimensions match
    if stride == 1 and in_channels == filters:
        return Add()([shortcut, x])

    return x

def efficient_conv_block(x, filters, kernel_size=3, strides=1):
    """
    An efficient convolution block using SeparableConv2D for better parameter efficiency.
    SeparableConv2D properly supports regularization and is more efficient than standard Conv2D.
    """
    x = SeparableConv2D(
        filters,
        kernel_size,
        strides=strides,
        padding='same',
        use_bias=False,
        depthwise_regularizer=l2(WEIGHT_DECAY),
        pointwise_regularizer=l2(WEIGHT_DECAY)
    )(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    return x

def build_efficient_chess_model(input_shape=(8, 8, 12), extra_features_shape=(13,)):
    """
    Efficient chess evaluation model using lightweight architecture.
    Optimized for low parameter count and fast inference.
    """
    board_input = Input(shape=input_shape, name='board')

    # Initial convolution - extract basic pattern features
    x = Conv2D(
        32, (3, 3),
        strides=(1, 1),
        padding='same',
        use_bias=False,
        kernel_regularizer=l2(WEIGHT_DECAY)
    )(board_input)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Use a mix of regular and separable convolutions for efficiency
    # Block 1
    shortcut1 = x
    x = efficient_conv_block(x, 32)
    x = efficient_conv_block(x, 32)
    # Add residual connection if shapes match
    if K.int_shape(shortcut1)[-1] == 32:
        x = Add()([x, shortcut1])

    # Block 2 - modest increase in filters
    shortcut2 = Conv2D(48, (1, 1), padding='same', use_bias=False)(x)
    shortcut2 = BatchNormalization()(shortcut2)
    x = efficient_conv_block(x, 48)
    x = efficient_conv_block(x, 48)
    x = Add()([x, shortcut2])

    # Block 3 - final feature extraction
    shortcut3 = Conv2D(64, (1, 1), padding='same', use_bias=False)(x)
    shortcut3 = BatchNormalization()(shortcut3)
    x = efficient_conv_block(x, 64)
    x = efficient_conv_block(x, 64)
    x = Add()([x, shortcut3])

    # Global pooling to reduce spatial dimensions
    x = GlobalAveragePooling2D()(x)

    # Process extra features separately with a small network
    extra_input = Input(shape=extra_features_shape, name='extra_features')
    extra_features = Dense(16, activation='relu', kernel_regularizer=l2(WEIGHT_DECAY))(extra_input)

    # Combine board features with extra features
    combined = Concatenate()([x, extra_features])

    # Fully connected layers with progressive reduction in size
    dense = Dense(48, activation='relu', kernel_regularizer=l2(WEIGHT_DECAY))(combined)
    dense = BatchNormalization()(dense)
    dense = Dense(24, activation='relu', kernel_regularizer=l2(WEIGHT_DECAY))(dense)

    # Output in range [-1, 1] using tanh
    # Using dtype='float32' to ensure consistent output type regardless of mixed precision
    output = Dense(1, activation='tanh', name='evaluation', dtype='float32')(dense)

    model = models.Model(inputs=[board_input, extra_input], outputs=output)
    return model

###########################################################
# 5. Quantization-Aware Training Helper Functions
###########################################################

def apply_quantization_to_dense(layer):
    """Apply quantization constraints to Dense layer weights."""
    return tf.quantization.fake_quant_with_min_max_args(
        layer, min=-6.0, max=6.0, num_bits=8
    )

def apply_quantization_to_conv(layer):
    """Apply quantization constraints to Conv layer weights."""
    return tf.quantization.fake_quant_with_min_max_args(
        layer, min=-6.0, max=6.0, num_bits=8
    )

# Note: For a full quantization-aware training approach, you would need to use
# TensorFlow Model Optimization Toolkit, which is beyond the simple example here.
# This is a simplified demonstration of the concept.

###########################################################
# 6. Training Data Setup
###########################################################

# Split into training/validation sets
val_size = max(1, int(len(tfrecord_files) * VALIDATION_SPLIT))
train_files = tfrecord_files[:-val_size]
val_files = tfrecord_files[-val_size:]

print(f"Training files: {len(train_files)}")
print(f"Validation files: {len(val_files)}")

# Quick sample to confirm dataset format
try:
    sample_dataset = create_dataset(train_files[0], batch_size=1)
    sample_batch = next(iter(sample_dataset))
    print("Sample shapes:")
    print(f"Board: {sample_batch[0]['board'].shape}, Extra: {sample_batch[0]['extra_features'].shape}")
    print(f"Eval: {sample_batch[1].shape}")

    # Now create the full datasets
    train_dataset = create_dataset(train_files, batch_size=BATCH_SIZE, is_training=True)
    val_dataset = create_dataset(val_files, batch_size=BATCH_SIZE, is_training=False)

    # Calculate steps_per_epoch based on number of samples
    # Estimate number of samples from number of files and batch size
    estimated_samples = len(train_files) * 10000  # Rough estimate, adjust based on your data
    steps_per_epoch = estimated_samples // BATCH_SIZE

except Exception as e:
    print(f"Error loading sample data: {str(e)}")
    raise

###########################################################
# 7. Training with Learning Rate Schedule and Callbacks
###########################################################

# Build the efficient model
efficient_model = build_efficient_chess_model(
    extra_features_shape=(sample_batch[0]['extra_features'].shape[1],)
)

# Use a fixed learning rate instead of a schedule since we'll use ReduceLROnPlateau
# This allows the callback to adjust the learning rate based on validation performance

# Compile with weighted huber loss
efficient_model.compile(
    optimizer=optimizers.Adam(learning_rate=INITIAL_LEARNING_RATE),
    loss=weighted_huber_loss(delta=0.1, focus_range=0.2, temp=8.0),
    metrics=['mae', 'mse']
)

print("\n=== Efficient Model Summary ===")
efficient_model.summary()

# Callbacks
model_callbacks = [
    callbacks.EarlyStopping(
        monitor='val_loss',
        patience=12,
        restore_best_weights=True,
        verbose=1
    ),
    callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=6,
        verbose=1,
        min_lr=1e-6
    ),
    callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/ChessEvaluation/efficient_model.keras',
        save_best_only=True,
        monitor='val_loss',
        verbose=1
    ),
    # TensorBoard callback for visualization if needed
    callbacks.TensorBoard(
        log_dir='/content/drive/MyDrive/ChessEvaluation/logs',
        histogram_freq=1,
        profile_batch=0  # Set to '0' to disable profiling to avoid slowdowns
    )
]

print("\n=== Training Efficient Model ===")
history = efficient_model.fit(
    train_dataset,
    epochs=EPOCHS,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_dataset,
    validation_steps=None,  # Calculate from dataset
    callbacks=model_callbacks,
    verbose=1
)

###########################################################
# 8. Model Pruning to Further Reduce Size
###########################################################

# Pruning requires tensorflow-model-optimization toolkit
try:
    import tensorflow_model_optimization as tfmot

    # Define pruning schedule
    pruning_params = {
        'pruning_schedule': tfmot.sparsity.keras.PolynomialDecay(
            initial_sparsity=0.0,
            final_sparsity=0.4,  # Target 40% sparsity (60% of weights remaining)
            begin_step=0,
            end_step=steps_per_epoch * 10  # Prune over 10 epochs
        )
    }

    print("\n=== Preparing Model for Pruning ===")
    # Clone the model before pruning to avoid errors
    model_for_pruning = keras.models.clone_model(efficient_model)
    model_for_pruning.set_weights(efficient_model.get_weights())

    # Apply pruning - wrap the entire model (correct way to handle Functional models)
    pruned_model = tfmot.sparsity.keras.prune_low_magnitude(
        model_for_pruning,
        **pruning_params
    )

    # Recompile the pruned model
    pruned_model.compile(
        optimizer=optimizers.Adam(learning_rate=INITIAL_LEARNING_RATE * 0.1),  # Lower learning rate for fine-tuning
        loss=weighted_huber_loss(delta=0.1, focus_range=0.2, temp=8.0),
        metrics=['mae', 'mse']
    )

    print("\n=== Pruned Model Summary ===")
    pruned_model.summary()

    # Fine-tune the pruned model with simpler callbacks
    pruning_callbacks = [
        callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True,
            verbose=1
        ),
        callbacks.ModelCheckpoint(
            '/content/drive/MyDrive/ChessEvaluation/efficient_pruned_model.keras',
            save_best_only=True,
            monitor='val_loss',
            verbose=1
        ),
        tfmot.sparsity.keras.UpdatePruningStep()
    ]

    print("\n=== Fine-tuning Pruned Model ===")
    pruning_history = pruned_model.fit(
        train_dataset,
        epochs=10,  # Short fine-tuning
        steps_per_epoch=steps_per_epoch,
        validation_data=val_dataset,
        callbacks=pruning_callbacks
    )

    # Strip pruning wrapper for final model
    final_model = tfmot.sparsity.keras.strip_pruning(pruned_model)

    # Save the pruned model
    final_model.save('/content/drive/MyDrive/ChessEvaluation/efficient_pruned_model.keras')

except ImportError:
    print("TensorFlow Model Optimization toolkit not found, skipping pruning step.")
    # Use the non-pruned model as the final model
    final_model = efficient_model
    final_model.save('/content/drive/MyDrive/ChessEvaluation/efficient_model.keras')
except Exception as e:
    print(f"Error during pruning: {str(e)}")
    print("Falling back to non-pruned model")
    # Use the non-pruned model as the final model
    final_model = efficient_model
    final_model.save('/content/drive/MyDrive/ChessEvaluation/efficient_model.keras')

###########################################################
# 9. Evaluation and Visualization
###########################################################

def plot_history(history, name="Efficient Chess Evaluation Model"):
    """Plot training and validation loss/MAE."""
    plt.figure(figsize=(15, 5))

    # Plot Loss
    plt.subplot(1, 3, 1)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title(f'{name} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Plot MAE
    plt.subplot(1, 3, 2)
    plt.plot(history.history['mae'], label='Train')
    plt.plot(history.history['val_mae'], label='Validation')
    plt.title(f'{name} MAE')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Absolute Error')
    plt.legend()
    plt.grid(True)

    # Plot MSE
    plt.subplot(1, 3, 3)
    plt.plot(history.history['mse'], label='Train')
    plt.plot(history.history['val_mse'], label='Validation')
    plt.title(f'{name} MSE')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Squared Error')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/ChessEvaluation/efficient_training_history.png')
    plt.show()

# Plot the training history
plot_history(history)

# Evaluate on a batch from validation set
test_batch = next(iter(val_dataset))
predictions = final_model.predict(test_batch[0])
true_evals = test_batch[1].numpy()

# Evaluate performance near zero
zero_threshold = 0.15
near_zero_indices = np.where(np.abs(true_evals) <= zero_threshold)[0]
print(f"\nPositions with ground-truth eval within ±{zero_threshold}: {len(near_zero_indices)}")

if len(near_zero_indices) > 0:
    mae_near_zero = np.mean(np.abs(predictions[near_zero_indices] - true_evals[near_zero_indices]))
    print(f"Efficient Model MAE near zero: {mae_near_zero:.4f}")

# Plot error distribution
plt.figure(figsize=(12, 5))

# Error distribution
plt.subplot(1, 2, 1)
errors = np.abs(predictions.flatten() - true_evals)
plt.hist(errors, bins=20, alpha=0.7)
plt.axvline(np.mean(errors), color='r', linestyle='dashed', linewidth=2)
plt.title('Error Distribution')
plt.xlabel('Absolute Error')
plt.ylabel('Count')
plt.grid(True)

# Error vs Evaluation Value
plt.subplot(1, 2, 2)
plt.scatter(true_evals, errors, alpha=0.5)
plt.xlabel('True Evaluation')
plt.ylabel('Absolute Error')
plt.title('Error vs Evaluation Value')
plt.grid(True)

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ChessEvaluation/efficient_error_distribution.png')
plt.show()

###########################################################
# 10. Export Model to ONNX Format
###########################################################
!pip install tf2onnx
import tf2onnx
import onnx

# Create an optimized version of the model for inference
def create_inference_model(model):
    """Create a model optimized for inference, with possible post-training optimizations."""
    # Clone the model architecture
    inference_model = tf.keras.models.clone_model(model)
    # Copy the weights
    inference_model.set_weights(model.get_weights())
    return inference_model

# Create inference model
inference_model = create_inference_model(final_model)

# Optional: Model weights rounding for better quantization
def round_weights(model, decimals=4):
    """Round weights to reduce precision requirements for inference."""
    weights = model.get_weights()
    rounded_weights = [np.round(w, decimals=decimals) for w in weights]
    model.set_weights(rounded_weights)
    return model

# Apply weight rounding (optional, remove if causes issues)
# inference_model = round_weights(inference_model, decimals=4)

# Define ONNX output path
onnx_path = "/content/drive/MyDrive/ChessEvaluation/efficient_chess_eval.onnx"

# Input specification
spec = (
    tf.TensorSpec((None, 8, 8, 12), tf.float32, name="board"),
    tf.TensorSpec((None, 13), tf.float32, name="extra_features")
)

# Convert to ONNX - use opset 13 for better compatibility
model_proto, _ = tf2onnx.convert.from_keras(inference_model, input_signature=spec, opset=13)

# Save the model
onnx.save(model_proto, onnx_path)
print(f"Efficient chess evaluation model saved to: {onnx_path}")

# Print model size information
import os
onnx_size_mb = os.path.getsize(onnx_path) / (1024 * 1024)
print(f"ONNX model size: {onnx_size_mb:.2f} MB")

# Final model summary
print("\n=== Final Model Summary ===")
print(f"- Total parameters: {final_model.count_params():,}")
print(f"- Final Validation Loss: {history.history['val_loss'][-1]:.4f}")
print(f"- Final Validation MAE: {history.history['val_mae'][-1]:.4f}")
print(f"- ONNX Model Size: {onnx_size_mb:.2f} MB")
```
