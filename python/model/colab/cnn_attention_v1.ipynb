```python
# Chess Evaluation Neural Network
# ==============================
# This notebook builds neural networks to evaluate chess positions from TFRecord data

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
from tensorflow.keras import layers, models, regularizers, optimizers, callbacks, mixed_precision
from tensorflow.keras.layers import Dense, Conv2D, BatchNormalization, Activation, Input, Add, Flatten
from tensorflow.keras.layers import GlobalAveragePooling2D, Concatenate, Reshape, MultiHeadAttention
from tensorflow.keras.regularizers import l2
from tensorflow.keras import backend as K

# Connect to Google Drive and set paths
from google.colab import drive
drive.mount('/content/drive')

# Path to your TFRecord files (with the correct path as specified)
TFRECORD_PATH = '/content/drive/MyDrive/ChessEvaluation/tfrecords'
mixed_precision.set_global_policy('mixed_float16')

tf.config.optimizer.set_jit(True)

# Define constants for training
BATCH_SIZE = 8192
EPOCHS = 30
VALIDATION_SPLIT = 0.1
LEARNING_RATE = 1e-3

# Check available TFRecord files
tfrecord_files = glob.glob(os.path.join(TFRECORD_PATH, '*.tfrecord'))
print(f"Found {len(tfrecord_files)} TFRecord files in {TFRECORD_PATH}")

# =========================================================
# 1. Data Loading Functions
# =========================================================

def parse_tfrecord_fn(example):
    """Parse a single TFRecord example into tensor features and labels."""
    # Update feature names to match your encoding
    feature_description = {
        'board_tensor': tf.io.FixedLenFeature([], tf.string),
        'extra_features': tf.io.FixedLenFeature([], tf.string),
        'evaluation': tf.io.FixedLenFeature([], tf.float32)
    }

    example = tf.io.parse_single_example(example, feature_description)

    # Parse the board representation (8x8x12 bitboard)
    board = tf.io.parse_tensor(example['board_tensor'], out_type=tf.float32)
    board = tf.reshape(board, [8, 8, 12])

    # Parse extra features (side to move, castling, en passant)
    extra = tf.io.parse_tensor(example['extra_features'], out_type=tf.float32)
    extra = tf.ensure_shape(extra, [13])  # Adjusted based on your encoding (1+4+8)

    # Get the evaluation label
    evaluation = example['evaluation']

    # Return in the format expected by the model
    return {'board': board, 'extra_features': extra}, evaluation

def create_dataset(file_pattern, batch_size=BATCH_SIZE, is_training=True):
    """Create a TensorFlow dataset from TFRecord files."""
    dataset = tf.data.TFRecordDataset(tf.io.gfile.glob(file_pattern))
    dataset = dataset.map(parse_tfrecord_fn, num_parallel_calls=tf.data.AUTOTUNE)

    if is_training:
        # Shuffle only during training
        dataset = dataset.shuffle(buffer_size=10000)

    # Batch and prefetch for better performance
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)

    return dataset

# =========================================================
# 2. Sample Weighting and Custom Loss Functions
# =========================================================

def create_sample_weights(labels, focus_range=0.15):
    """
    Create sample weights to emphasize positions near 0.

    Positions with evaluations in the range [-focus_range, focus_range]
    will get higher weights, with maximum weight at 0.
    """
    # Ensure labels are in the expected format
    labels = tf.cast(labels, tf.float32)

    # Create weights based on proximity to 0
    # The closer to 0, the higher the weight (inverse of absolute value)
    abs_values = tf.abs(labels)

    # Create a triangular weighting function that peaks at 0
    # and reaches 1.0 at Â±focus_range
    weights = 1.0 + (5.0 * tf.maximum(0.0, (focus_range - abs_values)) / focus_range)

    return weights

def weighted_mse(focus_range=0.15):
    """
    Custom loss function that gives higher weight to positions
    with evaluations close to 0.
    """
    def loss(y_true, y_pred):
        # Calculate standard MSE
        mse = tf.square(y_true - y_pred)

        # Apply weighting based on proximity to 0
        abs_values = tf.abs(y_true)
        weights = 1.0 + (5.0 * tf.maximum(0.0, (focus_range - abs_values)) / focus_range)

        # Return weighted MSE
        return tf.reduce_mean(mse * weights)

    return loss

def huber_zero_focused(delta=0.15, focus_range=0.15):
    """
    Custom Huber loss that gives higher weight to positions close to 0.
    Combines the robustness of Huber loss with emphasis on balanced positions.
    """
    def loss(y_true, y_pred):
        error = y_true - y_pred
        abs_error = tf.abs(error)

        # Standard Huber loss
        quadratic = tf.minimum(abs_error, delta)
        linear = abs_error - quadratic
        huber_loss = 0.5 * tf.square(quadratic) + delta * linear

        # Apply weighting based on proximity to 0
        abs_values = tf.abs(y_true)
        weights = 1.0 + (5.0 * tf.maximum(0.0, (focus_range - abs_values)) / focus_range)

        return tf.reduce_mean(huber_loss * weights)

    return loss

# =========================================================
# 3. Model A: Convolutional Neural Network (CNN) Architecture
# =========================================================

def build_cnn_model(input_shape=(8, 8, 12), extra_features_shape=(13,)):
    """
    Build a convolutional neural network for chess position evaluation.

    This model uses a series of convolutional layers to extract spatial patterns
    from the board representation, and combines them with the extra features
    before making the final evaluation.
    """
    # Board input
    board_input = Input(shape=input_shape, name='board')

    # Convolutional layers for the board
    x = Conv2D(64, (3, 3), padding='same', use_bias=False,
               kernel_regularizer=l2(1e-4))(board_input)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # More convolutional layers
    x = Conv2D(64, (3, 3), padding='same', use_bias=False,
               kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, (3, 3), padding='same', use_bias=False,
              kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Extract features
    x = Conv2D(128, (3, 3), padding='same', use_bias=False,
              kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Global features from the board
    x = GlobalAveragePooling2D()(x)

    # Extra features input (side to move, castling, en passant)
    extra_input = Input(shape=extra_features_shape, name='extra_features')

    # Combine board features and extra features
    combined = Concatenate()([x, extra_input])

    # Dense layers for evaluation
    dense = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(combined)
    dense = Dense(64, activation='relu', kernel_regularizer=l2(1e-4))(dense)
    dense = Dense(32, activation='relu', kernel_regularizer=l2(1e-4))(dense)

    # Output layer with tanh activation to ensure [-1, 1] range
    output = Dense(1, activation='tanh', name='evaluation', dtype='float32')(dense)

    # Create the model
    model = models.Model(inputs=[board_input, extra_input], outputs=output)

    return model

# =========================================================
# 4. Model B: Residual Network with Attention Mechanism
# =========================================================

def residual_block(x, filters, kernel_size=3):
    """A standard residual block for chess evaluation."""
    # Store the input for the residual connection
    residual = x

    # First convolutional layer
    y = Conv2D(filters, kernel_size, padding='same', use_bias=False,
              kernel_regularizer=l2(1e-4))(x)
    y = BatchNormalization()(y)
    y = Activation('relu')(y)

    # Second convolutional layer
    y = Conv2D(filters, kernel_size, padding='same', use_bias=False,
              kernel_regularizer=l2(1e-4))(y)
    y = BatchNormalization()(y)

    # Add the residual connection
    if residual.shape[-1] != y.shape[-1]:
        # If channels don't match, project residual
        residual = Conv2D(filters, (1, 1), padding='same', use_bias=False,
                         kernel_regularizer=l2(1e-4))(residual)
        residual = BatchNormalization()(residual)

    # Add the residual connection
    y = Add()([y, residual])
    y = Activation('relu')(y)

    return y

def attention_module(x, heads=4):
    """Apply self-attention to the chess board representation."""
    # Chess board dimensions are fixed
    h, w = 8, 8
    channels = x.shape[-1]

    # Reshape to sequence format for attention
    seq_length = h * w
    x_reshaped = Reshape((seq_length, channels))(x)

    # Apply multi-head self-attention
    attention_output = MultiHeadAttention(
        num_heads=heads, key_dim=channels // heads
    )(x_reshaped, x_reshaped)

    # Add residual connection and reshape back
    x_attended = Add()([x_reshaped, attention_output])
    output = Reshape((h, w, channels))(x_attended)

    return output

def build_resnet_attention_model(input_shape=(8, 8, 12), extra_features_shape=(13,), blocks=10):
    """
    Build a residual network with attention for chess position evaluation.

    This model combines residual blocks for feature extraction with
    self-attention mechanisms to capture long-range dependencies on the board.
    """
    # Board input
    board_input = Input(shape=input_shape, name='board')

    # Initial convolution
    x = Conv2D(64, (3, 3), padding='same', use_bias=False,
              kernel_regularizer=l2(1e-4))(board_input)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Residual blocks
    for block in range(blocks):
        filters = 128 if block < blocks // 2 else 256
        x = residual_block(x, filters)

        # Apply attention every other block after the first few blocks
        # Reducing the frequency of attention layers to prevent instability
        if block >= 2 and block % 2 == 0 and block < 6:  # Limit to early/middle layers
            x = attention_module(x, heads=2)  # Reduce number of heads

    # Global features from the board
    x = GlobalAveragePooling2D()(x)

    # Extra features input (side to move, castling, en passant)
    extra_input = Input(shape=extra_features_shape, name='extra_features')

    # Combine board features and extra features
    combined = Concatenate()([x, extra_input])

    # Dense layers for evaluation with reduced complexity
    dense = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(combined)
    dense = BatchNormalization()(dense)  # Add normalization to stabilize training
    dense = Dense(64, activation='relu', kernel_regularizer=l2(1e-4))(dense)
    dense = BatchNormalization()(dense)  # Add normalization to stabilize training

    # Output layer with tanh activation to ensure [-1, 1] range
    output = Dense(1, activation='tanh', name='evaluation', dtype='float32')(dense)

    # Create the model
    model = models.Model(inputs=[board_input, extra_input], outputs=output)

    return model

# Modified compilation settings to prevent NaN issues
def compile_resnet_model(model):
    """Compile the ResNet model with settings to prevent NaN values."""
    model.compile(
        optimizer=optimizers.Adam(
            learning_rate=5e-4,  # Lower learning rate for stability
            epsilon=1e-7,  # Increased epsilon to prevent division by zero
            clipnorm=1.0   # Gradient clipping to prevent exploding gradients
        ),
        loss=weighted_mse(focus_range=0.15),  # Using simpler loss function initially
        metrics=['mae']
    )
    return model

# Modified training function with callbacks to detect and handle NaNs
def train_with_nan_detection(model, train_dataset, val_dataset, model_path, epochs=20):
    """Train the model with NaN detection and prevention."""
    # Define callbacks
    callbacks_list = [
        # Stop training if NaN loss is encountered
        callbacks.TerminateOnNaN(),

        # Early stopping
        callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        ),

        # Learning rate reduction
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=3,
            min_lr=1e-6
        ),

        # Model checkpoint
        callbacks.ModelCheckpoint(
            filepath=model_path,
            monitor='val_loss',
            save_best_only=True
        )
    ]

    # Train with NaN detection
    history = model.fit(
        train_dataset,
        epochs=epochs,
        validation_data=val_dataset,
        callbacks=callbacks_list,
        verbose=1
    )

    return history


# =========================================================
# 5. Verify Dataset Format
# =========================================================

# Create training and validation datasets
train_files = tfrecord_files[:-max(1, int(len(tfrecord_files) * VALIDATION_SPLIT))]
val_files = tfrecord_files[-max(1, int(len(tfrecord_files) * VALIDATION_SPLIT)):]

print(f"Training files: {len(train_files)}")
print(f"Validation files: {len(val_files)}")

# Check the dataset format with a small sample
try:
    # Create a small sample dataset to check format
    sample_dataset = create_dataset(train_files[0], batch_size=1)

    # Get a sample batch
    print("Loading a sample batch to verify data format...")
    sample_batch = next(iter(sample_dataset))

    # Print shapes to verify
    print(f"Input shapes - Board: {sample_batch[0]['board'].shape}, Extra: {sample_batch[0]['extra_features'].shape}")
    print(f"Label shape: {sample_batch[1].shape}")
    print("Sample data loaded successfully!")

    # Now create the full datasets
    train_dataset = create_dataset(train_files, batch_size=BATCH_SIZE, is_training=True)
    val_dataset = create_dataset(val_files, batch_size=BATCH_SIZE, is_training=False)

except Exception as e:
    print(f"Error loading sample data: {str(e)}")

    # Print more diagnostic information
    print("\nDiagnostic information:")
    print(f"First few TFRecord files: {tfrecord_files[:3]}")

    # Try to examine the TFRecord format
    try:
        raw_dataset = tf.data.TFRecordDataset(tfrecord_files[0])
        raw_record = next(iter(raw_dataset))
        print("\nRaw TFRecord example (first few bytes):")
        print(raw_record.numpy()[:100])

        # Try inspecting with a different approach
        example = tf.train.Example()
        example.ParseFromString(raw_record.numpy())
        print("\nTFRecord features:")
        print(example.features.feature.keys())
    except Exception as nested_e:
        print(f"Error inspecting TFRecord format: {str(nested_e)}")

    print("\nPlease check your TFRecord format and update the parse_tfrecord_fn function accordingly.")
    raise

# =========================================================
# 6. Training Setup and Execution
# =========================================================

# Compile and train the CNN model
cnn_model = build_cnn_model(extra_features_shape=(sample_batch[0]['extra_features'].shape[1],))
cnn_model.compile(
    optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
    loss=weighted_mse(focus_range=0.15),
    metrics=['mae']
)

print("\n=== CNN Model Architecture ===")
cnn_model.summary()

# Define callbacks for training
cnn_callbacks = [
    callbacks.EarlyStopping(patience=10, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(factor=0.5, patience=5),
    callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/ChessEvaluation/cnn_model.keras',
        save_best_only=True
    )
]

# Train the CNN model
print("\n=== Training CNN Model ===")
cnn_history = cnn_model.fit(
    train_dataset,
    epochs=EPOCHS,
    validation_data=val_dataset,
    callbacks=cnn_callbacks
)

# Compile and train the ResNet+Attention model
resnet_attention_model = build_resnet_attention_model(
    extra_features_shape=(sample_batch[0]['extra_features'].shape[1],)
)
resnet_attention_model.compile(
    optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
    loss=huber_zero_focused(delta=0.1, focus_range=0.15),
    metrics=['mae']
)

print("\n=== ResNet+Attention Model Architecture ===")
resnet_attention_model.summary()

# Define callbacks for training
resnet_callbacks = [
    callbacks.EarlyStopping(patience=10, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(factor=0.5, patience=5),
    callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/ChessEvaluation/resnet_attention_model.keras',
        save_best_only=True
    )
]

# Train the ResNet+Attention model
print("\n=== Training ResNet+Attention Model ===")
resnet_history = resnet_attention_model.fit(
    train_dataset,
    epochs=EPOCHS,
    validation_data=val_dataset,
    callbacks=resnet_callbacks
)

# =========================================================
# 7. Evaluation and Visualization
# =========================================================

# Plot training history
def plot_history(histories, names):
    """Plot training and validation loss for multiple models."""
    plt.figure(figsize=(12, 5))

    # Plot loss
    plt.subplot(1, 2, 1)
    for hist, name in zip(histories, names):
        plt.plot(hist.history['loss'], label=f'{name} Training')
        plt.plot(hist.history['val_loss'], label=f'{name} Validation')

    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)

    # Plot MAE
    plt.subplot(1, 2, 2)
    for hist, name in zip(histories, names):
        plt.plot(hist.history['mae'], label=f'{name} Training')
        plt.plot(hist.history['val_mae'], label=f'{name} Validation')

    plt.title('Model MAE')
    plt.ylabel('Mean Absolute Error')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/ChessEvaluation/training_history.png')
    plt.show()

# Plot the training history for both models
plot_history(
    [cnn_history, resnet_history],
    ['CNN', 'ResNet+Attention']
)

# Create a test dataset with a small number of examples for analysis
test_dataset = next(iter(val_dataset.take(1)))

# Run predictions on both models
cnn_preds = cnn_model.predict(test_dataset[0])
resnet_preds = resnet_attention_model.predict(test_dataset[0])
true_evals = test_dataset[1].numpy()

# Analyze predictions near zero
zero_threshold = 0.15
near_zero_indices = np.where(np.abs(true_evals) <= zero_threshold)[0]

print("\n=== Evaluation in Positions Near Zero ===")
print(f"Positions with eval in [-{zero_threshold}, {zero_threshold}]: {len(near_zero_indices)}")

if len(near_zero_indices) > 0:
    cnn_mae_near_zero = np.mean(np.abs(cnn_preds[near_zero_indices] - true_evals[near_zero_indices]))
    resnet_mae_near_zero = np.mean(np.abs(resnet_preds[near_zero_indices] - true_evals[near_zero_indices]))

    print(f"CNN MAE near zero: {cnn_mae_near_zero:.4f}")
    print(f"ResNet+Attention MAE near zero: {resnet_mae_near_zero:.4f}")

# Plot the error distribution
plt.figure(figsize=(12, 5))

# CNN error distribution
plt.subplot(1, 2, 1)
cnn_errors = np.abs(cnn_preds.flatten() - true_evals)
plt.hist(cnn_errors, bins=20, alpha=0.7)
plt.axvline(np.mean(cnn_errors), color='r', linestyle='dashed', linewidth=2)
plt.title('CNN Error Distribution')
plt.xlabel('Absolute Error')
plt.ylabel('Count')
plt.grid(True)

# ResNet error distribution
plt.subplot(1, 2, 2)
resnet_errors = np.abs(resnet_preds.flatten() - true_evals)
plt.hist(resnet_errors, bins=20, alpha=0.7)
plt.axvline(np.mean(resnet_errors), color='r', linestyle='dashed', linewidth=2)
plt.title('ResNet+Attention Error Distribution')
plt.xlabel('Absolute Error')
plt.ylabel('Count')
plt.grid(True)

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ChessEvaluation/error_distribution.png')
plt.show()

# =========================================================
# 8. Model Comparison and Insights
# =========================================================

print("\n=== Model Comparison ===")
print("CNN Model:")
print(f"- Parameters: {cnn_model.count_params():,}")
print(f"- Final validation loss: {cnn_history.history['val_loss'][-1]:.4f}")
print(f"- Final validation MAE: {cnn_history.history['val_mae'][-1]:.4f}")

print("\nResNet+Attention Model:")
print(f"- Parameters: {resnet_attention_model.count_params():,}")
print(f"- Final validation loss: {resnet_history.history['val_loss'][-1]:.4f}")
print(f"- Final validation MAE: {resnet_history.history['val_mae'][-1]:.4f}")

# =========================================================
# 9. Export Models to ONNX Format
# =========================================================
!pip install tf2onnx

import tf2onnx
import onnx

# Define ONNX output paths
cnn_onnx_path = "/content/drive/MyDrive/ChessEvaluation/cnn_model.onnx"
resnet_onnx_path = "/content/drive/MyDrive/ChessEvaluation/resnet_attention_model.onnx"

# Input specification (same for both models)
spec = (
    tf.TensorSpec((None, 8, 8, 12), tf.float32, name="board"),
    tf.TensorSpec((None, 13), tf.float32, name="extra_features")
)

# Convert CNN model to ONNX
cnn_model_proto, _ = tf2onnx.convert.from_keras(cnn_model, input_signature=spec, opset=13)
onnx.save(cnn_model_proto, cnn_onnx_path)
print(f"CNN model saved to: {cnn_onnx_path}")

# Convert ResNet+Attention model to ONNX
resnet_model_proto, _ = tf2onnx.convert.from_keras(resnet_attention_model, input_signature=spec, opset=13)
onnx.save(resnet_model_proto, resnet_onnx_path)
print(f"ResNet+Attention model saved to: {resnet_onnx_path}")
```
